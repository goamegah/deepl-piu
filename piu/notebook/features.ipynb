```python
# Load datasets
train_df = pd.read_csv(f"{DATASET_PATH}/train.csv")
test_df = pd.read_csv(f"{DATASET_PATH}/test.csv")
```

```python
from piu.data.data_preprocessor import DataPreprocessor
```

```python
print(train_df.shape)
train_df.head()
```

```python
tr_df = train_df.copy()
tr_df = tr_df.drop(['id', 'PCIAT-PCIAT_Total'], axis=1)
tr_df.head()
```

```python
# Initialiser le prétraitement
preprocessor = DataPreprocessor(
    tr_df,
    target_column='sii',
    feature_selection_method="k_best",
    k_best=10,
    imputation_method="knn",
    drop_missing_target=True
)
    
# Transformer les données
X, y, class_weights = preprocessor.fit_transform()
```
